Fermi-Dirac distribution function

```python
import numpy as np
import matplotlib.pyplot as plt

# Define constants
k_B = 8.617333262145e-5 # Boltzmann constant in eV/K
T = 300  # Temperature in Kelvin (adjust as needed)
E = np.linspace(0, 0.2, 10000)  # Energy levels in electron volts (eV)

# Normalize energy levels
E -= np.min(E)

# Define the Fermi-Dirac distribution function
def fermi_dirac(energy, temperature):
    kT = k_B * temperature
    return 1 / (np.exp(energy / kT) + 1)

# Calculate the distribution
distribution = fermi_dirac(E, T)

# Create the plot
plt.figure(figsize=(8, 6))
plt.plot(E, distribution, label=f'Temperature = {T} K')
plt.xlabel('Normalized Energy (eV)')
plt.ylabel('Fermi-Dirac Distribution')
plt.title('Fermi-Dirac Distribution Function with Normalized Energy Values')
plt.legend()
plt.grid()
plt.show()
```

Bose-Einstein statistics distribution

```python
import numpy as np
import matplotlib.pyplot as plt

# Define constants
k_B = 8.617333262145e-5  # Boltzmann constant in eV/K
T = 300  # Temperature in Kelvin (adjust as needed)
E = np.linspace(1e-6, 2, 1000)  # Energy levels in electron volts (eV) with a small offset

# Define the Bose-Einstein distribution function
def bose_einstein(energy, temperature):
    kT = k_B * temperature
    denominator = np.exp(energy / kT) - 1
    return 1 / np.where(denominator != 0, denominator, 1e-6)  # Replace 0 with a small value

# Calculate the distribution
distribution = bose_einstein(E, T)

# Create the plot
plt.figure(figsize=(8, 6))
plt.plot(E, distribution, label=f'Temperature = {T} K')
plt.xlabel('Energy (eV)')
plt.ylabel('Bose-Einstein Distribution')
plt.title('Bose-Einstein Distribution Function')
plt.legend()
plt.grid()
plt.show()
```

Newton's laws of motion

```python
import numpy as np
import matplotlib.pyplot as plt

class Particle:
    def __init__(self, mass, initial_position, initial_velocity):
        self.mass = mass
        self.position = initial_position
        self.velocity = initial_velocity

    def apply_force(self, force):
        # Update velocity based on the force and mass
        acceleration = force / self.mass
        self.velocity += acceleration

    def update_position(self, time_step):
        # Update position based on the velocity and time step
        self.position += self.velocity * time_step

# Get user input for parameters
mass = float(input("Enter the mass of the particle (kg): "))
initial_position = float(input("Enter the initial position (meters): "))
initial_velocity = float(input("Enter the initial velocity (m/s): "))
time_step = float(input("Enter the time step for simulation (seconds): "))
total_time = float(input("Enter the total simulation time (seconds): "))

# Create a particle
particle = Particle(mass, initial_position, initial_velocity)

# Lists to store time, position, and velocity for plotting
time_values = []
position_values = []

# Simulate the motion
time = 0.0
while time <= total_time:
    # Get user input for force at each time step
    force = float(input(f"Enter the force at time {time:.2f} s (Newtons): "))

    # Update the particle's velocity and position
    particle.apply_force(force)
    particle.update_position(time_step)

    # Store time, position values for plotting
    time_values.append(time)
    position_values.append(particle.position)

    # Increment time
    time += time_step

# Plot the motion
plt.figure(figsize=(8, 6))
plt.plot(time_values, position_values)
plt.xlabel('Time (s)')
plt.ylabel('Position (m)')
plt.title('Particle Motion with User-Defined Force')
plt.grid()
plt.show()
```
